```python

import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

Load data

```python
spd_data = np.load("cleandata.npy", allow_pickle=True)
```

Find the cluster number

1.The elbow method

2.The optimization of the silhouette coefficient

3.The gap statistic

```python
def optimal_cluster_number(data, max_clusters):
    silhouette_scores = []
    for n_clusters in range(2, max_clusters+1):
        kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(data)
        silhouette_avg = silhouette_score(data, kmeans.labels_)
        silhouette_scores.append(silhouette_avg)
    return np.argmax(silhouette_scores) + 2
```

```python
# 假设您的数据已存储在名为data的(471, 1494) numpy数组中
best_k = optimal_cluster_number(spd_data.T, 10)  # 您可以根据需要更改max_clusters参数
kmeans = KMeans(n_clusters=best_k, random_state=0).fit(spd_data.T)
```

```python
clusters_pca = []
for i in range(best_k):
    cluster_data = spd_data[:, kmeans.labels_ == i]
    pca = PCA(n_components=8)
    pca.fit(cluster_data.T)
    mean_spectrum = cluster_data.mean(axis=1)
    clusters_pca.append({'pca': pca, 'mean_spectrum': mean_spectrum})
```

```python
data = np.array(clusters_pca)
```
