```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
spd_data = np.load("/Users/dongtianchi/Documents/GIT/SpectralReconstruction/ComputationalSpectrometers/Deeplearning/Main/denoised_data.npy", allow_pickle=True)
spd_data.shape
```

```python
negative_values = np.where(spd_data < 0)
band_indices, sample_indices = negative_values

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(sample_indices, band_indices, marker='o', c='red', edgecolors='black', s=50)
plt.xlabel("Sample Index")
plt.ylabel("Band Index")
plt.title("Negative Values in SPD Data")
plt.grid(True)
plt.show()
```

```python
sns.set_theme()
negative_values_df = pd.DataFrame({'Sample Index': sample_indices, 'Band Index': band_indices})
# 使用seaborn绘制散点图
plt.figure(figsize=(10, 6))
sns.scatterplot(data=negative_values_df, x='Sample Index', y='Band Index', color='red', edgecolor='black', s=50)
plt.grid(True)
plt.title("Negative Values in SPD Data")
plt.show()
```
